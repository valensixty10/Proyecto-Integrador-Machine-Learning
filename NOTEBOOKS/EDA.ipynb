## **Análisis Exploratorio de Datos (EDA)**
El Análisis Exploratorio de Datos (EDA) nos permite comprender mejor el dataset antes de aplicar modelos predictivos. A través del EDA, identificamos patrones, relaciones entre variables y posibles problemas, como valores nulos o outliers, que podrían afectar el modelado. En este análisis, inspeccionaremos las estadísticas descriptivas, las relaciones entre variables y visualizaremos la distribución de las variables clave.


### Inspección Inicial del Dataset
Cargamos el dataset y mostramos las primeras filas para obtener una vista preliminar de la estructura de los datos. Esta inspección nos permite identificar las variables presentes en el dataset y verificar que los datos se han cargado correctamente.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar el dataset
df = pd.read_csv(r"C:/Users/Usuario/Desktop/PROYECTO_M6/Data/ML_cars.csv")

# 2. Mostrar las primeras filas del dataset
print("Primeras 5 filas del dataset:")
df.head()
```

### Estadísticas Descriptivas
Las estadísticas descriptivas proporcionan un resumen básico de las variables numéricas en el dataset. A través de medidas como la media, la desviación estándar y los percentiles, obtenemos información clave sobre la distribución de los datos y detectamos posibles outliers o valores inusuales.

```python

# 3. Resumen de las estadísticas descriptivas
print("\nEstadísticas descriptivas:")
df.describe()
```

### Verificación de Valores Nulos
Es importante detectar la presencia de valores nulos en las columnas del dataset, ya que estos pueden afectar el desempeño de los modelos predictivos. Al conocer la cantidad de valores faltantes, podremos decidir si rellenarlos, eliminarlos o aplicar alguna otra técnica para manejarlos en fases posteriores.

```python
# 4. Verificar valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())
```

### Análisis de Variables Categóricas
Las variables categóricas requieren un análisis diferente al de las variables numéricas. Aquí analizamos la frecuencia de cada categoría para entender la distribución de los valores en columnas no numéricas, como el tipo de combustible o la carrocería. Esta información será útil para la fase de preparación de datos, donde transformaremos estas variables en formato adecuado para el modelado.


```python
# 5. Análisis de variables categóricas
categorical_cols = df.select_dtypes(include=['object']).columns
print("\nVariables categóricas y su distribución:")
for col in categorical_cols:
    print(f"\nFrecuencia de {col}:")
    print(df[col].value_counts())
```

### Distribución del Precio de los Vehículos
Visualizamos la distribución de la variable objetivo `price` mediante un histograma. Esta visualización nos ayuda a entender cómo se distribuyen los precios de los vehículos en el dataset, y a detectar si la variable presenta sesgos o outliers que puedan requerir tratamiento adicional.


```python
# 6. Visualización: Histograma de precios
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], bins=20, kde=True)
plt.title('Distribución del Precio de los Vehículos')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()
```

### Matriz de Correlación
La matriz de correlación nos permite observar la relación entre las variables numéricas del dataset. A través de un heatmap, podemos visualizar qué variables están más fuertemente correlacionadas entre sí, lo cual es útil para identificar posibles variables predictoras del precio. Un valor de correlación cercano a 1 o -1 indica una relación fuerte, mientras que valores cercanos a 0 indican poca o ninguna relación.


```python
# 7. Visualización: Matriz de correlación para variables numéricas
numeric_df = df.select_dtypes(include=['float64', 'int64'])  # Filtrar solo columnas numéricas
corr_matrix = numeric_df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()
```

Estos insights iniciales nos proporcionan una base sólida para la preparación de datos y el modelado predictivo en las siguientes fases del proyecto.
